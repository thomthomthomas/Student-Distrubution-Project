# üß† Abstraction of the Problem

To design a scalable and dynamic solution, we needed to abstract the problem to eliminate hard-coded constraints and handle edge cases. Here's how we approached it:

### üé≤ Initial Challenges
1. Gender Ratio Issues:  
   - Early groups depleted members of one gender, leaving later groups unbalanced.  
   - Fixed ratios didn't work due to outliers, e.g., tutorial groups with >30 members of one gender.

2. GPA Balancing Challenges:  
   - Pairing the highest and lowest CGPAs disrupted gender diversity.  
   - GPA balancing needed to dynamically adjust to maintain fairness.

### üîç Dynamic Abstraction
To address these challenges, we developed a two-pass system:
1. Pass 1: Populate groups with gender-balanced pairs.  
   - Ensures diversity in the core of each group.
   - Adjusts dynamically to handle tutorial groups with gender imbalances.
   
2. Pass 2: Fill remaining spots with median GPA students.  
   - Balances academic performance without disrupting gender diversity.

### üõ† Why This Works
- Scalability:  
  - Handles edge cases like gender-dominated tutorial groups or GPA outliers without manual adjustments.  
  - Extends to groups of sizes beyond 5 (e.g., 4‚Äì9 members).  
- Flexibility:  
  - Pairs (even) and leftovers (odd) allow seamless configuration for various group sizes.  
- Abstraction:  
  - Instead of hardcoding edge cases, we focused on designing a systematic, reusable approach.

### Section 1: Analyzing and Organizing Student Data

**Purpose**: Process a CSV file (`records.csv`) to group students by **Tutorial Groups (TG)** for analysis.

---

**Key Steps**:
1. **Read CSV**: Extract student details (ID, School, Name, Gender, GPA).
2. **Group by TG**: Organize students into a dictionary:
   - **Key**: Tutorial Group.
   - **Value**: List of student records.

---
**Example Output**:
{
    1: [{'ID': '1234', 'Sch': 'EEE', 'Name': 'Abi', 'Gender': 'Male', 'GPA': 3.8}, ... , ...],
    2: [{'ID': '5678', 'Sch': 'SSS', 'Name': 'Thomas', 'Gender': 'Male', 'GPA': 4.2} , ... , ...]
}

```python

import csv

# Define a function to analyze and process the contents of a CSV file
def analyse_csv():
    data = {}  # Initialize an empty dictionary to store data organized by tutorial group

    # Open the CSV file in read mode
    with open("assets/records.csv", "r") as file:
        csv_reader = csv.reader(file)  # Create a CSV reader object to read each line as a list
        next(csv_reader)  # Skip the header row

        # Loop through each row in the CSV file
        for row in csv_reader:
            # Extract the tutorial group number from the first element in the row
            # `split('-')[1]` extracts the numeric part of the group identifier
            tut_grp = int(row[0].split('-')[1])

            # Create a dictionary for this student's details
            student = {
                'ID': row[1],          # Student's unique ID
                'Sch': row[2],         # School abbreviation
                'Name': row[3],        # Full name of the student
                'Gender': row[4],      # Gender
                'GPA': float(row[5])   # GPA, converted to float
            }

            # If this tutorial group doesn't exist in `data`, initialize it as an empty list
            if tut_grp not in data:
                data[tut_grp] = []

            # Append the student's details to the list for their tutorial group
            data[tut_grp].append(student)

    # Return the dictionary containing organized data
    # Each key is a tutorial group, and each value is a list of student dictionaries
    return data

```

- **Purpose:** Identifies a suitable student from another group for swapping to balance school diversity.
- **Key Points:**
  - Matches based on:
    - Different school.
    - Same gender.
    - Similar GPA (within 0.2 difference).

---

## **6. `balance_teams`**
- **Purpose:** Balances overrepresented schools in groups by swapping students between groups.
- **Key Points:**
  - Identifies overrepresented schools (more than 2 students).
  - Swaps students to reduce overrepresentation while maintaining diversity.

### **Purpose of Helper Functions:**
- These functions are for implementing constraints, maintaining balance, and ensuring diversity during the allocation of students to project groups.

     

```python
def merge_sort(gendered_list):
    """
    Sort students by GPA using the merge sort algorithm.

    Parameters:
        gendered_list (list): A list of student dictionaries containing GPA as a key.

    Returns:
        list: A list of students sorted in descending order of GPA.
    """
    # Base case: A list with 0 or 1 elements is already sorted.
    if len(gendered_list) <= 1:
        return gendered_list

    # Split the list into two halves.
    median = len(gendered_list) // 2
    left = merge_sort(gendered_list[:median])  # Recursively sort the left half.
    right = merge_sort(gendered_list[median:])  # Recursively sort the right half.

    # Merge the two sorted halves.
    sorted_list = []
    while left and right:
        # Compare GPAs to determine the order.
        if left[0]['GPA'] > right[0]['GPA']:
            sorted_list.append(left.pop(0))  # Add the student with the higher GPA.
        else:
            sorted_list.append(right.pop(0))  # Add the other student.

    # Add any remaining elements from the left and right lists.
    sorted_list.extend(left)
    sorted_list.extend(right)

    return sorted_list


def can_add_pair_to_group(pair, school_counts):
    """
    Check if adding this pair to a group maintains school diversity constraints.

    Parameters:
        pair (list): A pair of student dictionaries to evaluate.
        school_counts (dict): Tracks the number of students from each school in a group.

    Returns:
        bool: True if the pair can be added while maintaining constraints, False otherwise.
    """
    # Check each student in the pair for school diversity constraints.
    for student in pair:
        school = student['Sch']  # Get the student's school.
        if school_counts.get(school, 0) >= 2:  # Max 2 students from the same school allowed.
            return False

    # Additional constraint: If both students in the pair are from the same school, max 1 is allowed.
    if pair[0]['Sch'] == pair[1]['Sch']:
        school = pair[0]['Sch']
        if school_counts.get(school, 0) >= 1:
            return False

    return True


def can_add_student_to_group(student, school_counts):
    """
    Check if adding this student to a group maintains school diversity constraints.

    Parameters:
        student (dict): The student dictionary to evaluate.
        school_counts (dict): Tracks the number of students from each school in a group.

    Returns:
        bool: True if the student can be added while maintaining constraints, False otherwise.
    """
    school = student['Sch']  # Get the student's school.
    if school_counts.get(school, 0) >= 2:  # Max 2 students from the same school allowed.
        return False
    return True


def update_school_counts(school_counts, students):
    """
    Update the school counts for a group after adding students.

    Parameters:
        school_counts (dict): Tracks the number of students from each school in a group.
        students (list): A list of students added to the group.

    Returns:
        None
    """
    for student in students:
        school = student['Sch']  # Get the student's school.
        # Increment the count for the school.
        school_counts[school] = school_counts.get(school, 0) + 1


def find_swap_candidate(current_team, student, project_groups):
    """
    Find a valid swap candidate in other groups to balance school diversity.

    Parameters:
        current_team (list): The group where the student is currently assigned.
        student (dict): The student to find a swap candidate for.
        project_groups (dict): All project groups.

    Returns:
        tuple: A tuple containing the target group and swap candidate, or (None, None) if no candidate is found.
    """
    target_gender = student['Gender']
    target_gpa = student['GPA']
    target_school = student['Sch']

    # Iterate through other teams to find a valid swap candidate.
    for team_name, team in project_groups.items():
        if team is current_team:  # Skip the current team.
            continue
        for other_student in team:
            # Check for gender match, school diversity, and GPA similarity.
            if (other_student['Sch'] != target_school and
                other_student['Gender'] == target_gender and
                abs(other_student['GPA'] - target_gpa) <= 0.2):
                return team, other_student  # Found a valid candidate.
    return None, None  # No suitable candidate found.


def balance_teams(project_groups):
    """
    Balance school representation across project groups by swapping students.

    Parameters:
        project_groups (dict): All project groups.

    Returns:
        dict: Updated project groups with improved school diversity.
    """
    swaps_made = True
    while swaps_made:
        swaps_made = False
        for team_name, team in project_groups.items():
            school_counts = {}
            # Count the number of students from each school in the team.
            for student in team:
                school = student['Sch']
                school_counts[school] = school_counts.get(school, 0) + 1

            # Identify overrepresented schools (more than 2 students).
            overrepresented_schools = [school for school, count in school_counts.items() if count >= 3]

            # Attempt to swap students from overrepresented schools.
            for school in overrepresented_schools:
                student_to_swap = next((student for student in team if student['Sch'] == school), None)
                if not student_to_swap:
                    continue

                # Find a valid swap candidate.
                target_team, swap_candidate = find_swap_candidate(team, student_to_swap, project_groups)
                if swap_candidate:
                    # Perform the swap between teams.
                    team.remove(student_to_swap)
                    target_team.append(student_to_swap)
                    target_team.remove(swap_candidate)
                    team.append(swap_candidate)

                    print(f"Swapped: {student_to_swap['Name']} <--> {swap_candidate['Name']}")
                    swaps_made = True
                    break  # Restart the loop after a swap to recheck constraints.
            if swaps_made:
                break

    return project_groups
```

### Section 3: Forming balanced Groups

---

## **Objective**
To allocate students into balanced project groups for each tutorial group, ensuring diversity in GPA, gender, and school representation.

---

## **Steps in the Main Grouping Function**

### **1. Separate Students by Gender**
- **Purpose:** Enables balanced gender representation in pairs and groups.
- **Key Process:** Split students into two lists: men and women.

---

### **2. Sort Students by GPA**
- **Purpose:** Ensures academically balanced pairing and group distribution.
- **Key Process:**
  - Sort men in descending GPA order.
  - Sort women in ascending GPA order.

---

### **3. Form Pairs**
- **Purpose:** Create gender-diverse and academically balanced pairs.
- **Key Process:**
  - Match the smartest man with the least smart woman and vice versa.
  - Stop when:
    - Either list is exhausted.
    - The maximum number of pairs (20) is formed.

---

### **4. Distribute Pairs Across Groups**
- **Purpose:** Assign pairs to project groups while adhering to constraints.
- **Key Constraints:**
  - Max 4 students per group for pairs.
  - Max 2 students from the same school per group.
- **Key Process:**
  - Use a round-robin approach to distribute pairs across groups.
  - Relax constraints if no group can accommodate a pair.

---

### **5. Add Leftover Students**
- **Purpose:** Ensure all students are assigned to groups.
- **Key Constraints:**
  - Max 5 students per group.
  - Max 2 students from the same school per group.
- **Key Process:**
  - Use a round-robin approach to assign leftovers to groups.
  - Relax constraints if no group can accommodate a student.

---

### **6. Balance Teams**
- **Purpose:** Reduce overrepresentation of schools within groups.
- **Key Process:**
  - Identify overrepresented schools (more than 2 students).
  - Swap students between groups to address imbalances.

---

### **7. Store Results**
- **Purpose:** Save the finalized project groups for each tutorial group.
- **Key Process:** Add completed groups to the `updated` dictionary for further analysis or export.

---

## **Key Features**
- Ensures diversity in groups based on GPA, gender, and school representation.
- Provides flexibility by relaxing constraints when necessary.
- Balances groups post-allocation to maintain diversity.


```python
# Analyze the CSV data and initialize the results dictionary.
data = analyse_csv()
updated = {}

# Iterate through each tutorial group and its list of students.
for tutorial_group, students in data.items():
    # Separate students by gender for balanced pairing.
    men = [student for student in students if student['Gender'] == 'Male']
    women = [student for student in students if student['Gender'] == 'Female']

    # Sort men in descending GPA order and women in ascending GPA order.
    men = merge_sort(men)
    women = merge_sort(women)

    # Initialize project groups (A-J) and school count trackers for this tutorial group.
    project_groups = {group: [] for group in "ABCDEFGHIJ"}
    school_counts = {group: {} for group in "ABCDEFGHIJ"}

    # Form pairs of students for balanced gender and academic performance.
    pairs = []
    leftovers = []  # Track students who cannot be paired.

    # Match students: Smartest man with least smart woman, and vice versa.
    while len(men) > 1 and len(women) > 1 and len(pairs) < 20:
        pair1 = [men.pop(0), women.pop(-1)]  # Smartest man with least smart woman.
        pair2 = [men.pop(-1), women.pop(0)]  # Least smart man with smartest woman.
        pairs.append(pair1)
        pairs.append(pair2)

    # Add any remaining men or women to the leftovers list.
    leftovers.extend(men + women)

    # Distribute pairs across project groups.
    group_index = 0  # Start with the first group.
    for pair in pairs:
        added = False
        attempts = 0  # Keep track of attempts to distribute the pair.
        while not added and attempts < 10:  # Try assigning to all 10 groups.
            group_name = "ABCDEFGHIJ"[group_index]  # Get the current group name.
            # Check group constraints: max 4 students per group and school diversity.
            if len(project_groups[group_name]) < 4 and can_add_pair_to_group(pair, school_counts[group_name]):
                # Assign the pair to the group and update school counts.
                project_groups[group_name].extend(pair)
                update_school_counts(school_counts[group_name], pair)
                added = True
            group_index = (group_index + 1) % 10  # Move to the next group.
            attempts += 1

        # Relax constraints if no group could accommodate the pair.
        if not added:
            print(f"Relaxing criteria for {tutorial_group}")
            for group_name in "ABCDEFGHIJ":
                if len(project_groups[group_name]) < 4:
                    project_groups[group_name].extend(pair)
                    update_school_counts(school_counts[group_name], pair)
                    break

    # Add leftover students to project groups.
    group_index = 0  # Start with the first group again.
    for student in leftovers:
        added = False
        attempts = 0  # Keep track of attempts to distribute the student.
        while not added and attempts < 10:  # Try assigning to all 10 groups.
            group_name = "ABCDEFGHIJ"[group_index]
            # Check group constraints: max 5 students per group and school diversity.
            if len(project_groups[group_name]) < 5 and can_add_student_to_group(student, school_counts[group_name]):
                # Assign the student to the group and update school counts.
                project_groups[group_name].append(student)
                update_school_counts(school_counts[group_name], [student])
                added = True
            group_index = (group_index + 1) % 10  # Move to the next group.
            attempts += 1

        # Relax constraints if no group could accommodate the student.
        if not added:
            print(f"Relaxing criteria for {tutorial_group}")
            for group_name in "ABCDEFGHIJ":
                if len(project_groups[group_name]) < 5:
                    project_groups[group_name].append(student)
                    update_school_counts(school_counts[group_name], [student])
                    break

    # Perform balancing to address overrepresented schools in groups.
    project_groups = balance_teams(project_groups)

    # Store the finalized project groups for this tutorial group.
    updated[tutorial_group] = project_groups
```

### SUMMARY/PRESENTATION FINDINGS
## **Objective**
To evaluate project group diversity by calculating scores based on GPA balance, gender ratio, and school representation, and measure the overall diversity performance.

---

## **1. Calculate Tutorial Group Average GPAs**
- **Function:** `calculate_tutorial_group_averages(updated)`
- **Purpose:** Computes the average GPA for each tutorial group.
- **Key Process:**
  - Extract GPAs from all groups within a tutorial group.
  - Calculate the mean GPA for each tutorial group.
- **Output:** Dictionary of average GPAs for each tutorial group.

---

## **2. Calculate Diversity Scores**
- **Function:** `calculate_diversity_scores(updated, tutorial_group_averages)`
- **Purpose:** Assigns a diversity score to each project group based on three metrics:
  1. **GPA Balance:**
     - Compares the group‚Äôs average GPA to the tutorial group‚Äôs average GPA.
     - Closer alignment results in higher scores (up to 50 points).
  2. **Gender Balance:**
     - Rewards groups with a balanced male-to-female ratio (up to 30 points).
  3. **School Diversity:**
     - Scores based on the distribution of students from different schools (up to 20 points).
- **Output:** Updates the `diversity_scores` dictionary with total scores for each group.

---

## **3. Calculate Average Diversity Score**
- **Function:** `calculate_average_diversity_score(diversity_scores)`
- **Purpose:** Calculates the mean diversity score across all project groups.
- **Key Process:**
  - Sum all diversity scores.
  - Divide by the number of project groups.
- **Output:** A single average diversity score.

---

## **Performance Metrics**
1. **Lowest Diversity Score:**
   - Identifies the group with the weakest diversity score.
2. **Perfect Scores:**
   - Counts groups with a perfect score (100).
3. **High Scores:**
   - Counts groups with scores ‚â• 95.

---

## **Key Results**
- Displays diversity scores for all groups.
- Highlights overall diversity performance metrics, including perfect and near-perfect scores.

---

## **Key Features**
- **GPA Alignment:** Ensures balanced academic performance.
- **Gender Balance:** Rewards gender diversity.
- **School Representation:** Promotes inclusion of multiple schools in groups.

```python
# Initialize a dictionary to store diversity scores for all project groups.

diversity_scores = {}

# Initialize a list to store GPA differences for all project groups.
gpa_differences = []

# Initialize counters for gender ratios.
total_male_count = 0
total_female_count = 0
total_groups = 0

# Function to calculate the average GPA for each tutorial group.
def calculate_tutorial_group_averages(updated):
    """
    Calculate the average GPA of students in each tutorial group.

    Parameters:
        updated (dict): A dictionary of tutorial groups and their assigned project groups.

    Returns:
        dict: A dictionary containing the average GPA for each tutorial group.
    """
    tutorial_group_averages = {}
    for tg, groups in updated.items():
        # Extract all GPAs from all groups in the tutorial group.
        all_gpas = [student['GPA'] for group in groups.values() for student in group]
        # Calculate the average GPA if there are students in the group.
        tutorial_group_averages[tg] = sum(all_gpas) / len(all_gpas) if all_gpas else 0
    return tutorial_group_averages

# Function to calculate diversity scores for all project groups.
def calculate_diversity_scores(updated, tutorial_group_averages):
    """
    Calculate diversity scores for each project group based on GPA, gender, and school representation.

    Parameters:
        updated (dict): A dictionary of tutorial groups and their assigned project groups.
        tutorial_group_averages (dict): Average GPA for each tutorial group.

    Returns:
        None: Updates the global `diversity_scores` dictionary.
    """
    global total_male_count, total_female_count, total_groups
    for tg, groups in updated.items():
        tutorial_avg_gpa = tutorial_group_averages[tg]  # Retrieve the average GPA for the tutorial group.
        for group_label, students in groups.items():
            # Calculate GPA score.
            group_avg_gpa = sum(student['GPA'] for student in students) / len(students)
            gpa_difference = abs(group_avg_gpa - tutorial_avg_gpa)
            gpa_differences.append(gpa_difference)  # Add to the global list for final analysis.
            if gpa_difference < 0.01:
                gpa_score = 50
            elif gpa_difference < 0.025:
                gpa_score = 45
            elif gpa_difference < 0.05:
                gpa_score = 40
            elif gpa_difference < 0.075:
                gpa_score = 35
            elif gpa_difference < 0.1:
                gpa_score = 30
            elif gpa_difference < 0.2:
                gpa_score = 25
            else:
                gpa_score = 0

            # Calculate gender score.
            male_count = sum(1 for student in students if student['Gender'] == 'Male')
            female_count = len(students) - male_count
            total_male_count += male_count
            total_female_count += female_count
            total_groups += 1
            gender_ratio = (male_count, female_count)
            if gender_ratio in [(3, 2), (2, 3)]:
                gender_score = 30
            elif gender_ratio in [(4, 1), (1, 4)]:
                gender_score = 15
            else:
                gender_score = 0

            # Calculate school diversity score.
            school_counts = {}
            for student in students:
                school = student['Sch']
                school_counts[school] = school_counts.get(school, 0) + 1
            school_distribution = sorted(school_counts.values(), reverse=True)
            if len(school_distribution) == 5:
                school_score = 20
            elif len(school_distribution) == 4:
                school_score = 17
            elif school_distribution == [2, 2, 1]:
                school_score = 15
            elif school_distribution == [3, 1, 1]:
                school_score = 10
            elif school_distribution == [3, 2]:
                school_score = 8
            elif school_distribution == [4, 1]:
                school_score = 5
            else:
                school_score = 0

            # Calculate the total diversity score for the group.
            total_score = gpa_score + gender_score + school_score
            group_full_label = f"{tg}{group_label}"  # Example: '1A', '1B', ...
            diversity_scores[group_full_label] = total_score

# Function to calculate the average diversity score across all groups.
def calculate_average_diversity_score(diversity_scores):
    """
    Calculate the average diversity score for all project groups.

    Parameters:
        diversity_scores (dict): A dictionary of diversity scores for each project group.

    Returns:
        float: The average diversity score.
    """
    total_score = sum(diversity_scores.values())
    average_score = total_score / len(diversity_scores) if diversity_scores else 0
    return average_score

# Step 1: Calculate the average GPA for each tutorial group.
tutorial_group_averages = calculate_tutorial_group_averages(updated)

# Step 2: Calculate diversity scores for all project groups.
calculate_diversity_scores(updated, tutorial_group_averages)

# Step 3: Calculate and display the average diversity score.
average_diversity_score = calculate_average_diversity_score(diversity_scores)
print("Diversity Scores for Each Group:", diversity_scores)
print("Average Diversity Score:", average_diversity_score)

# Display additional performance metrics.
print(min(diversity_scores.values()), "is the lowest diversity score")

# Count the number of groups with perfect or near-perfect scores.
perfect_score_count = sum(1 for score in diversity_scores.values() if score == 100)
high_score_count = sum(1 for score in diversity_scores.values() if score >= 95)
print(f"{perfect_score_count} groups have a perfect score\n{high_score_count} groups have a score above 95")

# Calculate mean GPA difference.
mean_gpa_difference = sum(gpa_differences) / len(gpa_differences)

# Calculate median GPA difference.
sorted_gpa_differences = sorted(gpa_differences)
n = len(sorted_gpa_differences)
median_gpa_difference = sorted_gpa_differences[n // 2]

# Calculate average Male:Female ratio.
average_gender_ratio = (total_male_count / total_groups, total_female_count / total_groups)

print(f"Mean GPA Difference Across All Groups: {mean_gpa_difference}")
print(f"Median GPA Difference Across All Groups: {median_gpa_difference}")
print(f"Average Male:Female Ratio Across All Groups: {average_gender_ratio[0]:.2f}:{average_gender_ratio[1]:.2f}")
```

# üöß Limitations of Our Approach

Our solution effectively addresses the problem, but certain limitations reflect the trade-offs between scalability, simplicity, and fairness. Here's an overview:

### 1Ô∏è‚É£ Inherent Lack of Gender Diversity in Tutorial Groups

### 2Ô∏è‚É£ School Diversity Swapping
- To achieve school diversity, we implemented a swapping algorithm:
  - The algorithm identifies groups with imbalanced school representation and attempts to swap members between groups.
  - This process has a time complexity of O(n¬≤), which becomes inefficient as the dataset grows.
- Challenges:
  - The algorithm relies on hardcoded school limits, which may not reflect the unique composition of each tutorial group.
  - For this reason, we restrict the swapping functionality to groups of 5, ensuring manageable complexity.

---

### 3Ô∏è‚É£ CGPA Balancing
Balancing CGPA presents nuanced challenges:
- Observed Issue:  
  High-performing students deviate further from the mean than low-performing students. This causes some groups' average CGPA to be higher than the tutorial group's mean.
  
- Possible Fixes:
  1. Pair multiple low-CGPA students with the highest performers:
     - Drawback: Disrupts our pair-based system, reducing scalability.
  2. Assign lower-CGPA students as leftovers to groups with higher-than-average CGPA:
     - Drawback: Requires calculating group averages after each assignment, reducing efficiency.
  3. Implement a separate swapping algorithm:
     - Swap students with similar gender and school but differing CGPAs.
     - Drawback: Swapping introduces inefficiencies and can unintentionally reverse previous optimizations.

---

### üåü Balancing Simplicity, Scalability, and Results
Given the complexity of these challenges, we consciously accepted these limitations to maintain a balance between:
- Scalability: Avoiding excessive computational costs.
- Simplicity: Ensuring the algorithm remains modular and understandable.
- Fair Results: Achieving reasonable diversity without over-complicating the solution.

```python
import csv


def analyse_csv():
  data={}
  with open("records.csv","r") as file:
    csv_reader=csv.reader(file)
    next(csv_reader)  #first row of the csv is just the headings
    for row in csv_reader:
      tut_grp=int(row[0].split('-')[1])    # Changing the tutorial group name from G-1 to just 1 (maybe it will help us with sorting or something later)
      student={
          'ID':row[1],
          'Sch':row[2],
          'Name':row[3],
          'Gender':row[4],
          'GPA':float(row[5])}
      if tut_grp not in data:
                data[tut_grp] = []  # Initialize the list if group doesn't exist
      data[tut_grp].append(student)  # Add student to the group list
  return data

def merge_sort(gendered_list):
  # Sort students by GPA using merge sort.
    if len(gendered_list) <= 1:
        return gendered_list

    # Split the list into two halves
    median = len(gendered_list) // 2
    left = merge_sort(gendered_list[:median])
    right = merge_sort(gendered_list[median:])

    # Merge the two sorted halves
    sorted_list = []
    while left and right:
        if left[0]['GPA'] > right[0]['GPA']:  # Compare GPA
            sorted_list.append(left.pop(0))
        else:
            sorted_list.append(right.pop(0))

    # Add any remaining elements
    sorted_list.extend(left)
    sorted_list.extend(right)

    return sorted_list


# Helper functions used in the main logic
def can_add_pair_to_group(pair, school_counts):
    """ Check if adding this pair to the group maintains school diversity. """
    for student in pair:
        school = student['Sch']
        if school_counts.get(school, 0) >= 4:
            return False  # Would exceed school count limit
    if pair[0]['Sch'] == pair[1]['Sch']:
        school = pair[0]['Sch']
        if school_counts.get(school,0) >= 3:
            return False

    return True

def can_add_student_to_group(student, school_counts):
    """ Check if adding this student to the group maintains school diversity. """
    school = student['Sch']
    if school_counts.get(school, 0) >= 4:
        return False  # Would exceed school count limit
    return True

def update_school_counts(school_counts, students):
    """ Update the school counts for the group after adding students. """
    for student in students:
        school = student['Sch']
        school_counts[school] = school_counts.get(school, 0) + 1

def assign_teams(students, team_sizes):
    # Prepare teams based on provided sizes
    teams = {chr(65 + i): [] for i in range(len(team_sizes))}  # Teams A, B, C, ...
    school_counts = {chr(65 + i): {} for i in range(len(team_sizes))}  # Track school counts per team
    mens = [student for student in students if student['Gender'] == 'Male']
    womens = [student for student in students if student['Gender'] == 'Female']

    # Separate students by gender and sort by GPA
    men=merge_sort(mens)
    women=merge_sort(womens)

    # Create pairs
    pairs = []
    while len(men) > 1 and len(women) > 1 and len(pairs) < 20:
        pairs.append([men.pop(0), women.pop(-1)])  # Smartest man with least smart woman
        pairs.append([men.pop(-1), women.pop(0)])  # Least smart man with smartest woman

    # Leftover students
    leftovers = men + women
    # Round-robin adding pairs to teams
    team_index = 0
    for pair in pairs:
        added = False
        attempts = 0
        while not added and attempts < len(team_sizes):
            team_name = chr(65 + team_index)  # Get team label
            if len(teams[team_name]) + len(pair) <= team_sizes[team_index] and can_add_pair_to_group(pair, school_counts[team_name]):
                teams[team_name].extend(pair)
                update_school_counts(school_counts[team_name], pair)
                added = True
            team_index = (team_index + 1) % len(team_sizes)
            attempts += 1

        # Relax criteria if no group could accommodate the pair
        if not added:
            print(f"Relaxing criteria for team {tutorial_group}")
            for group_name in teams.keys():
                if len(teams[group_name]) + len(pair) <= team_sizes[ord(group_name) - 65]:
                    teams[group_name].extend(pair)
                    update_school_counts(school_counts[group_name], pair)
                    break

    # Round-robin adding leftovers to teams
    team_index = 0
    for student in leftovers:
        added = False
        attempts = 0
        while not added and attempts < len(team_sizes):
            team_name = chr(65 + team_index)
            if len(teams[team_name]) < team_sizes[team_index] and can_add_student_to_group(student, school_counts[team_name]):
                teams[team_name].append(student)
                update_school_counts(school_counts[team_name], [student])
                added = True
            team_index = (team_index + 1) % len(team_sizes)
            attempts += 1

        # Relax criteria if no group could accommodate the student
        if not added:
            print(f"Relaxing criteria for team {tutorial_group}")
            for group_name in teams.keys():
                if len(teams[group_name]) < team_sizes[ord(group_name) - 65]:
                    teams[group_name].append(student)
                    update_school_counts(school_counts[group_name], [student])
                    break

    return teams


def create_group_sizes(total_students):
    current_list = []
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    index = 0

    while total_students > 0:
        print(f"Students remaining: {total_students}")
        user_input = input(f"Enter Group {alphabet[index]} Size: ")

        # Reset command
        if user_input.lower() == "reset":
            print("Resetting the process...\n")
            total_students = 50
            current_list = []
            index = 0
            continue

        # Check if input is a valid integer
        try:
            group_size = int(user_input)
        except ValueError:
            print("Error: Invalid input. Please enter a number between 4 and 9, or 'reset' to restart.\n")
            continue

        # Check if input is within the valid range (4-9)
        if group_size < 4 or group_size > 9:
            print("Error: Out of Range. Please enter a number between 4 and 9.\n")
            continue

        # Check if the input group size can be added without causing remaining students to drop below 4 (unless it's the final group)
        if group_size > total_students:
            print(f"Error: Group size bigger than remaining.\n")
        elif total_students - group_size < 4 and total_students != group_size:
            print("Error: Insufficient remaining students for another group if this size is accepted.\n")
        else:
            # Accept the group size and update counters
            current_list.append(group_size)
            total_students -= group_size
            index += 1
            print(f"Current List: {current_list}\n")

    print("Thank you for submitting Group sizes")
    print("Your result is", current_list)
    final_count_list=current_list[:]
    return final_count_list




final_count_list=create_group_sizes(50)
data=analyse_csv()
updated={}

for tutorial_group, students in data.items():
    teams = assign_teams(students, final_count_list)

    # Store the result in the updated dictionary with the tutorial group as the key
    updated[tutorial_group] = teams


with open("Extra_Credit.csv", mode='w', newline='') as file:
    writer = csv.writer(file)

    # Write the header row
    header = ['Tutorial Group', 'Project Group', 'Student ID', 'School', 'Name', 'Gender', 'CGPA']
    writer.writerow(header)

    # Iterate through the dictionary and write rows
    for TG, groups in updated.items():
        for group_label, members in groups.items():
            for member in members:
                # Prepare the row with the relevant information
                row = [TG, group_label, member['ID'], member['Sch'], member['Name'], member['Gender'], member['GPA']]
                writer.writerow(row)

```
